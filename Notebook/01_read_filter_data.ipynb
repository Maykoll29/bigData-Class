## Definición de cada columna

Este dataset usa términos estándar de Darwin Core/GBIF. Las definiciones son generales para cada campo del archivo.

1. `gbifID` - Identificador único asignado por GBIF a cada registro de ocurrencia.
2. `datasetKey` - Identificador del conjunto de datos o publicación a la que pertenece el registro.
3. `occurrenceID` - Identificador único de la ocurrencia en el origen de datos, normalmente URL o clave.
4. `kingdom` - Reino biológico al que pertenece el organismo (por ejemplo, Plantae).
5. `phylum` - Filo o división taxonómica del organismo.
6. `class` - Clase taxonómica del organismo.
7. `order` - Orden taxonómico del organismo.
8. `family` - Familia taxonómica del organismo.
9. `genus` - Género taxonómico del organismo.
10. `species` - Especie científica a nivel de nombre binomial.
11. `infraspecificEpithet` - Epiteto infraespecífico cuando aplica (subespecie, variedad, forma, etc.).
12. `taxonRank` - Nivel taxonómico del nombre registrado (SPECIES, GENUS, etc.).
13. `scientificName` - Nombre científico completo del taxón.
14. `verbatimScientificName` - Nombre científico tal como aparece en la fuente original, sin normalizar.
15. `verbatimScientificNameAuthorship` - Autoría original del nombre científico tal como aparece en el dato fuente.
16. `countryCode` - Código de país en formato ISO 3166-1 alpha-2 donde se observó el organismo.
17. `locality` - Descripción textual del lugar específico de la ocurrencia.
18. `stateProvince` - Estado, provincia o unidad subnacional donde se registró la ocurrencia.
19. `occurrenceStatus` - Estado de la ocurrencia, por ejemplo PRESENT, ABSENT.
20. `individualCount` - Número de individuos observados o muestreados en ese registro.
21. `publishingOrgKey` - Identificador de la organización que publicó el registro.
22. `decimalLatitude` - Latitud en coordenadas decimales del punto de ocurrencia.
23. `decimalLongitude` - Longitud en coordenadas decimales del punto de ocurrencia.
24. `coordinateUncertaintyInMeters` - Incertidumbre de las coordenadas en metros.
25. `coordinatePrecision` - Precisión de las coordenadas.
26. `elevation` - Elevación o altitud del punto de ocurrencia, normalmente en metros.
27. `elevationAccuracy` - Precisión o exactitud de la elevación.
28. `depth` - Profundidad del punto de ocurrencia, relevante en registros acuáticos.
29. `depthAccuracy` - Precisión de la profundidad medida.
30. `eventDate` - Fecha del evento de observación o recolección en formato ISO 8601.
31. `day` - Día del mes extraído de la fecha del evento.
32. `month` - Mes extraído de la fecha del evento.
33. `year` - Año extraído de la fecha del evento.
34. `taxonKey` - Clave numérica interna de GBIF para el taxón.
35. `speciesKey` - Clave numérica interna de GBIF para la especie.
36. `basisOfRecord` - Tipo de registro o evidencia (p.ej. PRESERVED_SPECIMEN, HUMAN_OBSERVATION).
37. `institutionCode` - Código de la institución que conserva el registro o espécimen.
38. `collectionCode` - Código de la colección específica dentro de la institución.
39. `catalogNumber` - Número de catálogo o etiqueta del espécimen/registro.
40. `recordNumber` - Número de registro asignado en la fuente original.
41. `identifiedBy` - Nombre de la persona que identificó el espécimen u observación.
42. `dateIdentified` - Fecha en que se hizo la identificación.
43. `license` - Tipo de licencia de uso de los datos (por ejemplo CC_BY_4_0).
44. `rightsHolder` - Persona u organización que conserva los derechos sobre el registro.
45. `recordedBy` - Nombre de la persona que recolectó o registró la observación.
46. `typeStatus` - Estatus del tipo nomenclatural del espécimen, si aplica (p.ej. holotype, paratype).
47. `establishmentMeans` - Origen de la población en ese lugar (native, introduced, naturalized, etc.).
48. `lastInterpreted` - Fecha en que GBIF o la plataforma interpretó/actualizó el registro.
49. `mediaType` - Tipo de material asociado al registro, como imágenes, sonido, etc.
50. `issue` - Indicador de problemas o cuestiones de calidad/datos detectados en el registro.


Cada fila es un registro de ocurrencia biológica: un organismo observado o un espécimen recolectado.

### Qué identifica la fila como única
- `gbifID`: es el identificador único asignado por GBIF a ese registro. Normalmente es la clave principal para distinguir filas en este dataset.
- `occurrenceID`: es el identificador original de la ocurrencia en la fuente de datos. También debe ser único dentro del origen original.
- `datasetKey`: identifica el conjunto de datos o publicación al que pertenece esa fila, pero no identifica la fila por sí sola.

### Qué significa cada fila
Cada fila describe, para esa ocurrencia:
- el taxón del organismo (`kingdom`, `phylum`, `class`, `order`, `family`, `genus`, `species`, etc.)
- dónde y cuándo se registró (`countryCode`, `locality`, `stateProvince`, `decimalLatitude`, `decimalLongitude`, `eventDate`, `year`, `month`, `day`)
- qué tipo de evidencia es (`basisOfRecord`)
- metadatos del registro y del espécimen (`institutionCode`, `collectionCode`, `catalogNumber`, `recordNumber`, `identifiedBy`, `dateIdentified`)
- licencia y derechos (`license`, `rightsHolder`)
- calidad y estado del dato (`occurrenceStatus`, `coordinateUncertaintyInMeters`, `issue`, `establishmentMeans`)

### Resumen
- Una fila = un evento/ocurrencia biológica.
- `gbifID` es la llave más clara para saber que la fila es única.
- `occurrenceID` es la referencia original de la ocurrencia.
- `datasetKey` dice de qué dataset viene.

## Spark: preparar datos clave para zonas calientes de biodiversidad

A partir de todo el CSV, esta celda conserva solo las columnas necesarias para analizar latitud, longitud, especie y contar registros por punto. El resultado es un dataset más ligero y listo para identificar hotspots.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when
import os

csv_path = os.path.join(os.getcwd(), '0004764-260519110011954.csv')
print('Procesando CSV con Spark desde', csv_path)

spark = SparkSession.builder \
    .appName("BiodiversidadColombia") \
    .master("local[*]") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

required_cols = ["species", "decimalLatitude", "decimalLongitude"]
raw_df = spark.read \
    .option("header", "true") \
    .option("sep", "\t") \
    .option("inferSchema", "false") \
    .csv(csv_path)

missing_cols = [c for c in required_cols if c not in raw_df.columns]
if missing_cols:
    raise ValueError(f"Faltan columnas en el CSV: {missing_cols}. Columnas disponibles: {raw_df.columns}")

print('Columnas disponibles:', raw_df.columns)

df = raw_df.select(*required_cols) \
    .withColumn("decimalLatitude", col("decimalLatitude").cast("double")) \
    .withColumn("decimalLongitude", col("decimalLongitude").cast("double")) \
    .withColumn("species", when(col("species").isNull() | (col("species") == ""), "unknown").otherwise(col("species")))

clean_df = df.filter(
    col("decimalLatitude").isNotNull() &
    col("decimalLongitude").isNotNull() &
    (col("decimalLatitude") >= -90) & (col("decimalLatitude") <= 90) &
    (col("decimalLongitude") >= -180) & (col("decimalLongitude") <= 180)
)

print('Registros totales:', df.count())
print('Registros con coordenadas válidas:', clean_df.count())

hotspot_df = clean_df.groupBy("species", "decimalLatitude", "decimalLongitude").count().orderBy(col("count").desc())
print('Top 20 combinaciones de especie + coordenadas (cantidad):')
hotspot_df.show(20, truncate=False)

species_counts = clean_df.groupBy("species").count().orderBy(col("count").desc())
print('Top 20 especies por cantidad de registros:')
species_counts.show(20, truncate=False)

clean_out = os.path.join(os.getcwd(), 'biodiversidad_limpa.parquet')
hotspot_out = os.path.join(os.getcwd(), 'biodiversidad_especies_por_punto.parquet')
clean_df.write.mode("overwrite").parquet(clean_out)
hotspot_df.write.mode("overwrite").parquet(hotspot_out)

print('Guardado Parquet de datos limpios:', clean_out)
print('Guardado Parquet de especie+coordenadas+cantidad:', hotspot_out)

spark.stop()
```

## Rejilla geográfica para el mapa: `lat | long | especie | cantidad`

Esta celda crea una tabla por cuadro geográfico usando la latitud y longitud redondeadas. Cada fila representa un cuadrante, la especie más abundante en ese cuadro y su cantidad.

- `lat` y `long`: coordenadas del centro del cuadro geométrico.
- `especie`: la especie dominante en ese cuadro.
- `cantidad`: número de registros de esa especie en el cuadro.

Si quieres más detalles, cambia `round(..., 1)` a `round(..., 2)` para cuadrantes más chicos.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, round, desc, row_number
from pyspark.sql.window import Window
import os

spark = SparkSession.builder \
    .appName("BiodiversidadColombiaGrid") \
    .master("local[*]") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

clean_parquet = os.path.join(os.getcwd(), 'biodiversidad_limpa.parquet')
print('Leyendo datos limpios desde:', clean_parquet)
df = spark.read.parquet(clean_parquet)

# Generar el cuadro geográfico en Colombia
# 1 decimal = cuadrantes de tamaño regional
# 2 decimales = cuadros mucho más pequeños

df_grid = df \
    .withColumn('lat', round(col('decimalLatitude'), 1)) \
    .withColumn('long', round(col('decimalLongitude'), 1))

# Contar cuántos registros tiene cada especie por cuadro
conteo_por_cuadro = df_grid \
    .groupBy('lat', 'long', 'species') \
    .count() \
    .withColumnRenamed('count', 'cantidad')

# Elegir sólo la especie más abundante por cuadro
ventana_cuadro = Window.partitionBy('lat', 'long').orderBy(desc('cantidad'))

tabla_cuadros = conteo_por_cuadro \
    .withColumn('rank', row_number().over(ventana_cuadro)) \
    .filter(col('rank') == 1) \
    .drop('rank') \
    .select(
        col('lat'),
        col('long'),
        col('species').alias('especie'),
        col('cantidad')
    ) \
    .orderBy(desc('cantidad'))

print('Top 50 cuadros con las especies más abundantes:')
tabla_cuadros.show(50, truncate=False)

salida_parquet = os.path.join(os.getcwd(), 'datos_para_mapa_rejilla.parquet')
tabla_cuadros.write.mode('overwrite').parquet(salida_parquet)
print('Guardado el resumen de cuadriculado en:', salida_parquet)

spark.stop()
```
