```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("netflix-inc/netflix-prize-data")

print("Path to dataset files:", path)
```

```python
from pathlib import Path

root = Path(path)
text_files = sorted(root.rglob("*.txt"))
print("Archivos de texto disponibles:")
for file in text_files:
    print("-", file)

if text_files:
    file_to_read = text_files[0]
    print(f"\nLeyendo las primeras 20 líneas de: {file_to_read}\n")
    with file_to_read.open("r", encoding="latin-1", errors="replace") as f:
        for i, line in enumerate(f, start=1):
            print(line.rstrip())
            if i >= 20:
                break
else:
    print("No se encontró ningún archivo de texto en el dataset.")
```

```python
titles_file = Path(path) / "movie_titles.txt"
movie_id = 1234

with titles_file.open("r", encoding="latin-1", errors="replace") as f:
    for line in f:
        id_str, year, title = line.strip().split(",", 2)
        if id_str == str(movie_id):
            print("Película:", title)
            break
```
